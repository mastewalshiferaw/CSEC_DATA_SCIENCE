```python
# Loading Data
df = pd.read_csv('data/train.csv')

# Exploration
print("--- HEAD ---")
display(df.head())
print("\n--- INFO ---")
df.info()
print("\n--- DESCRIBE ---")
display(df.describe())

# Cleaning
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])
df = df.drop_duplicates()
```

```python
# Analysis
print("Survival by Gender:\n", df.groupby('Sex')['Survived'].mean())
print("\nSurvival by Class:\n", df.groupby('Pclass')['Survived'].mean())
print("\nAvg Age per Class:\n", df.groupby('Pclass')['Age'].mean())

# Age Groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])
print("\nSurvival by Age Group:\n", df.groupby('AgeGroup')['Survived'].mean())

# Filtering
survived_female = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
survived_children = df[(df['Age'] <= 12) & (df['Survived'] == 1)]
high_prob_1st_class = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
```
